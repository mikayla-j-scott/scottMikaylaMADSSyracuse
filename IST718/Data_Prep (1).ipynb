```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import FreqDist
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

## User Defined Functions

```python
#+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-
# Function: scrape_lyrics
# Purpose: Scrape lyrics from chartlyrics.com using the
#         artist name and the title of the song, save those
#         lyrics as a string and return them in the
#         variable named lyrics
#+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-
def scrape_lyrics(artist, song):
  chart_url = 'http://api.chartlyrics.com/apiv1.asmx/SearchLyricDirect'
  params = {'artist': artist, 'song': song}
  response = requests.get(chart_url, params=params)
  soup = BeautifulSoup(response.content, "xml")
  lyrics = soup.find("Lyric").text
  return lyrics
```

```python
#+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-
# Function: remove_stop
# Purpose: Remove stop words from lyrics
#+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-
def remove_stop(df):
    wordList = []
    sentenceList = []
    for x in df['tok_lyrics']:
        for y in x:
            if y not in stopwords:
                wordList.append(y)
        sentenceList.append(wordList)
        wordList = []
    df['stop_removed_lyrics'] = sentenceList
```

```python
# add in the missing lyrics
# added in excel, bring in the new file
# the new file is called lyrics_filled_in.csv

lyric_set = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IST718Final/lyrics_filled_in.csv')
lyric_set.head()
```

```python
import re

# lower case
loweredStrings = []
for x in lyric_set['lyrics']:
  loweredStrings.append(x.lower())
lyric_set['lower_lyrics'] = loweredStrings


# remove punctuation
prStrings = []
for x in lyric_set['lower_lyrics']:
  prStrings.append(re.sub(r'[^\w\s]', '', x))
lyric_set['pr_lower_lyrics'] = prStrings

# tokenize list
tokLists = []
for x in lyric_set['pr_lower_lyrics']:
  tokLists.append(x.split(" "))
lyric_set['tok_lyrics'] = tokLists


# count how many words before removal
wordCount = []
for x in lyric_set['tok_lyrics']:
  wordCount.append(len(x))
lyric_set['word_count'] = wordCount
```

```python
# bring in stop words
stopwords = list(nltk.corpus.stopwords.words("english"))
stopwords.append("")
x = ['ive', "really", 'come','thats' ,'didnt' ,'gonna' ,'hes'
,'his' ,'youre' ,'you' ,'im' ,'theyre' ,'hes' ,'oh','didnt' ,'thats'
,'dont','way' ,'got','yeah','okay','ill' ,'theres' ,'gotta','man', 'whats', 'ooh', 'ah', 'oh',
'yeah', 'yee', 'woah', 'wo']
for y in x:
    stopwords.append(y)
type(stopwords)
```

```python
# clean the lyrics
# remove stop words
remove_stop(lyric_set)

#back into a sentence
ListofSentences = []
for x in lyric_set['stop_removed_lyrics']:
    newSentence = ' '.join(x)
    ListofSentences.append(newSentence)
lyric_set['cleaned_lyrics'] = ListofSentences

# word count after removal
wordCount = []
for x in lyric_set['stop_removed_lyrics']:
  wordCount.append(len(x))
lyric_set['word_count_after_removal'] = wordCount
```

```python
# NLP (save neg, neu, cmpd, pos in cols in df)
#sentiment
res = []
for x in lyric_set['cleaned_lyrics']:
    polScores = SentimentIntensityAnalyzer().polarity_scores(x)
    res.append(polScores)

#concat df
resDF = pd.DataFrame(res)
lyric_set = pd.concat([lyric_set, resDF], axis=1)
```

```python
lyric_set['num_words_removed'] = lyric_set['word_count'] - lyric_set['word_count_after_removal']
```
